Here I create the creature class. It is initialised with a name and default values for health (hp), attack, defence, and speed. I also initialise the class with an additional health value, maxhp, to represent the maximum health points of a creature. When a new creature is created, a statement is printed identifying the creation and the creature details.

Three methods are created for this class, check_life, attack, and turn.

The check_life method allows us to check the current health points of a creature. If a creature's health points are above 0, the method returns the creature's current health points. If a creature's health points are below 0, this can happen if an attack causes damage greater than the remaining health points of a creature, the method returns hp = 0, signifying that the creature has died.

The attack method determines how a creature performs an attack on a specified target, and if it is successful. The attack begins by generating a random number between 1 and 20. If this number is greater than the targets combined defence and speed values, the attack will hit. If not, the attack is unsuccessful. If the attack hits, the attack damage is determined by the combination of the creatures attack value and a random number between 1 and 4. This value is subtracted from the target's health points. Print statements let us know if an attack is successful or not.

The turn method determines how a creature acts when it is their turn. In the case of a creature, it uses it's attack method.

The string magic method allows us to print the creature object.

```python
class Creature:

    def __init__(self, name, hp = 10, attack = 1, defence = 5, speed = 5, maxhp =10):
        self.name = name
        self.maxhp = maxhp
        self.hp = hp
        self.abilities = {"Attack": attack, "Defence": defence, "Speed": speed}
        print('New Creature created! ', self.__str__())

    def check_life(self):
        if self.hp > 0:
            return self.hp
        else:
            self.hp = 0
            return self.hp

    def attack(self, target):
        attackRoll = random.randint(1,20)
        targetDefence = int(target.abilities['Defence']) + int(target.abilities['Speed'])
        print('{} attacks {}.'.format(self.name, target.name))
        if attackRoll > targetDefence:
            attackRoll = int(self.abilities['Attack']) + random.randint(1,4)
            target.hp -= attackRoll
            print('Attack hits for {} damage!'.format(attackRoll))
        else:
            print("Attack not successful.")

    def turn(self, round_num, target):
        self.attack(target)
        
    def __str__(self):
        return "Name: {}, Health Points: {}, Attack: {}, Defence: {}, Speed: {}, Max HP: {}".format(self.name,self.hp,self.abilities['Attack'],self.abilities['Defence'],self.abilities['Speed'], self.maxhp)
```

Here I create two Creature objects, panda and bear. Using a for loop, I set the battle to last between 1 and 20 rounds. At the start of each round I print the round number and the current health of both creatures. Using the turn method, panda goes first and attacks bear. After each creature's turn, their opponent's life is checked. If the check_life method returns 0, the opponent has died, a message is printed identifying who has died, who the winner is and how many rounds it took. The program then breaks out of the loop.

```python
# Task 1 - Creature v Creature
print()
panda = Creature("Panda")
bear = Creature ("Bear")
print()

for i in range(1, 20):
    print('Round: {}'.format(i))
    print('Panda Health Points: {}'.format(panda.hp))
    print('Bear Health Points: {}'.format(bear.hp))
    print()
    panda.turn(i, bear)
    bearLife = bear.check_life()
    if bearLife == 0:
        print("\n{} died!".format(bear.name))
        print('{} is the winner. It took {} rounds.'.format(panda.name, i))
        break
    bear.turn(i, panda)
    pandaLife = panda.check_life()
    if pandaLife == 0:
        print("\n{} died!".format(panda.name))
        print('{} is the winner. It took {} rounds.'.format(bear.name, i))
        break
    print()
```

**Task 2**

Here I create the fighter class. A fighter is a type of creature and inherits the creature class. It is initialised in the same way as the creature class but with different default values. These values are used to initialise a creature when a fighter is created. A shield variable is created as a flag variable used in different methods.

As well as the creature methods, the fighter has a shield_up, shield_down, and a variation of the turn method.

When the shield_up variable is called, it first checks the value of the flag variable. If the flag variable value is 0, it updates the fighter's abilities, reducing it's attack capabilities by 5 and increasing it's defence capabilities by 5. It then changes the value of the flag variable, signifying that these changes were made. Finally, it prints a message to say that the fighter's abilities have changed.

When the shield_down variable is called, it acts in the same way as the shield_up variable, but returns the fighters abilities to their original values.

In each round (1-4), the fighter operates in different ways. In round 1 it attacks and then puts it's shield up. In round 2 and 3 it just attacks, and in round 4 it puts it's shield down and then attacks. The cycle then repeats.

For the fighter to operate correctly, we need to know what round it is. This is fine for rounds 1 to 4, but in round 5, the fighter needs to think it's round 1 again. To do this, if the round is greater than 4, we divide the round by 4, convert it to a string, and store the value in a variable. We then slice the string to get the decimal point and the numbers after it, convert it back to a float, and multiply it by 4.

Example: 
- Round = 5 (Really round 1)
- 5/4 = 1.25
- 4*.25 = **1**
-----
- Round = 7 (Really round 3)
- 7/4 = 1.75
- 4*.75 = **3**

This doesn't work for any rounds divisible by 4, but if the round is divisible by 4, we know it's the fourth round. In each fourth round we check the life of the opponent and print a message if they are still alive.

Using this method we can continue to change the fighter's actions throughout the fight. This method of determining the round is used in each of the remaining classes.

```python
# Task 2 Creature v Fighter

class Fighter(Creature):

    def __init__(self, name, hp = 50, attack = 5, defence = 10, speed = 3, maxhp = 50):
        Creature.__init__(self, name, hp, attack, defence, speed, maxhp)
        self.shield = 0

    def shield_up(self):
        if self.shield == 0:
            self.abilities['Attack'] -= 5
            self.abilities['Defence'] += 5
            self.shield = 1
            print('{} puts shields up! Attack decreased by 5 to {}. Defence increased by 5 to {}.'.format(self.name, self.abilities['Attack'], self.abilities['Defence']))
    
    def shield_down(self):
        if self.shield == 1:
            self.abilities['Attack'] += 5
            self.abilities['Defence'] -= 5
            self.shield = 0
            print('{} abilities returned to normal. Attack: {}. Defence: {}.'.format(self.name, self.abilities['Attack'], self.abilities['Defence']))
    
    def turn(self, round_num, target):
        if round_num > 4 and round_num % 4 != 0:
            whichRound = str(round_num/4)
            whichRound = whichRound[1:]
            whichRound = float(whichRound)
            round_num = 4*whichRound
        else:
            round_num == round_num

        if round_num == 1:
            self.attack(target)
            self.shield_up()
        elif round_num== 2 or round_num == 3:
            self.attack(target)
        if round_num % 4 == 0:
            self.shield_down()
            self.attack(target)
            if target.hp > 0:
                print('{} is still alive. Keep fighting!'.format(target.name))
```

Here I create a creature object, the salmon, and a fighter object, the polar bear, and set them to fight for between 1 and 20 rounds. This fighting process is identical to the prior creature v creature fighting process.

```python
print()
salmon = Creature('Salmon')
polarBear = Fighter('Polar Bear')
print()

for i in range(1, 20):
    print('Round: {}'.format(i))
    print('Salmon Health Points: {}'.format(salmon.hp))
    print('Polar Bear Health Points: {}'.format(polarBear.hp))
    print()
    salmon.turn(i, polarBear)
    polarLife = polarBear.check_life()
    if polarLife == 0:
        print("\n{} died!".format(polarBear.name))
        print('{} is the winner. It took {} rounds.'.format(salmon.name, i))
        break
    polarBear.turn(i, salmon)
    salmonLife = salmon.check_life()
    if salmonLife == 0:
        print("\n{} died!".format(salmon.name))
        print('{} is the winner. It took {} rounds.'.format(polarBear.name, i))
        break
    print()
```

Here I create the Archer class. An archer is a type of creature and inherits the creature class. It is initialised in the same way as the creature class but with different default values. These values are used to initialise a creature when an archer is created. An abilitiesChanged variable is created as a flag variable used in different methods.

As well as the creature methods, the archer has a sneak attack method and a variation of the attack and turn methods.

The sneak attack method is a different form of attack. To determine the attackRoll variable, the larger of 2 randomly generated numbers between 1 and 20 is taken. The archer's speed and the speed of the target are then compared. If the archer's speed is greater, the difference in speeds is added to the attack roll. If the flag variable is equal to 0, signifying that the archer's abilities are in their default state, the archer's abilities are updated, increasing the attack value by 3 and decreasing the defence value by 3. The flag variable is then changed to 1 to signify that changes have been made. The attack roll is then compared to a combination of the target's defence and speed values. If greater, the attack hits and the damage is determined by a combination of the archer's attack value and a randomly generated number between 1 and 8. Statements are printed identifying whether attack was successful or not.

The attack method works identical to the creature's attack method but the archer's abilities are returned to normal if they have been changed due to a previous sneak attack.

The turn method implements different actions depending on the round.

```python
# Task 3 Archer v Fighter

class Archer(Creature):

    def __init__(self, name, hp = 30, attack = 7, defence = 8, speed = 8, maxhp = 30):
        Creature.__init__(self, name, hp, attack, defence, speed, maxhp)
        self.abilitiesChanged = 0
        
    def sneak_attack(self, target):
        roll_nums = [random.randint(1,20), random.randint(1,20)]
        attackRoll = max(roll_nums)
        selfSpeed = int(self.abilities['Speed'])
        targetSpeed = int(target.abilities['Speed'])
        speedDiff = selfSpeed - targetSpeed
        if selfSpeed > targetSpeed:
            attackRoll += speedDiff

        if self.abilitiesChanged == 0:
            self.abilities['Attack'] += 3
            self.abilities['Defence'] -= 3
            self.abilitiesChanged = 1
            print('{} abilities updated. Attack increased by 3 to {}. Defence decreased by 3 to {}.'.format(self.name, self.abilities['Attack'], self.abilities['Defence']))

        targetDefence = int(target.abilities['Defence']) + int(target.abilities['Speed'])
        print('{} sneak attacks {}.'.format(self.name, target.name))
        if attackRoll > targetDefence:
            attackRoll = int(self.abilities['Attack']) + random.randint(1,8)
            target.hp -= attackRoll
            print('Attack hits for {} damage!'.format(attackRoll))
        else:
            print("Attack not successful.")

    def attack(self, target):
        if self.abilitiesChanged == 1:
            self.abilities['Attack'] -= 3
            self.abilities['Defence'] += 3
            self.abilitiesChanged = 0
            print('{} abilities returned to normal. Attack: {}. Defence: {}.'.format(self.name, self.abilities['Attack'], self.abilities['Defence']))

        Creature.attack(self, target)

    def turn(self, round_num, target):
        if round_num > 4 and round_num % 4 != 0:
            whichRound = str(round_num/4)
            whichRound = whichRound[1:]
            whichRound = float(whichRound)
            round_num = 4*whichRound
        else:
            round_num == round_num

        if round_num == 1:
            self.attack(target)
        elif round_num == 2 or round_num == 3:
            self.sneak_attack(target)
        if round_num % 4 == 0:
            self.sneak_attack(target)
            if target.hp > 0:
                print('{} is still alive. Keep fighting!'.format(target.name))
```

Here I create the Wizard class. A wizard is a type of creature and inherits the creature class. It is initialised in the same way as the creature class but with different default values. These values are used to initialise a creature when a wizard is created. The wizard also has a variable for mana and arcana.

It's string magic method is the same but adds the Mana and Arcana values.

In addition to the creature methods, the wizard class has methods for recharge, fire bolt, heal, mass heal, and fire storm. It also has a variation of the attack method.

The attack method is the same as the creature method but also tops up the wizards mana by 20. It only adds the whole 20 if the current mana plus 20 is less than or equal to the max mana of 100. Otherwise it tops up mana to the max of 100.

The recharge method tops up a wizard's mana points by 30. Like the attack method, it only adds the whole 30 if the current mana plus 30 is less than or equal to the max mana of 100. Otherwise it tops up mana to the max of 100. If mana is already at the maximum of 100, it prints a message saying mana is full.

The fire bolt method is a form of attack. A random number is generated from 1 to 20, determining the attack roll. If greater than the combination of the target's defence and speed, the attack hits. The damage is determined by a combination of the wizard's attack value, a randomly generated number from 1 to the wizard's current arcana value, and half of the wizard's current arcana value. If the attack hits, the wizard gains 10 mana. These 10 mana are applied with the same rules as before.

The heal method allows the wizard to heal another creature. The wizard needs at least 20 mana to perform a heal. The heal amount is determined by a random number between 1 and 8, plus half of the wizard's current arcana value. If the heal will push a creature's health points beyond their max health points, their health is just restored to maximum. If a wizard has insufficent mana, a message is printed identifying how much mana they are short.

The mass heal method works similar to the heal method but takes a list of creatures as an argument. Rather than healing one creature, it heals all of the creatures in the list. The heal amount is determined by a random number between 1 and 10, plus the wizard's total current arcana value. The cost of mass healing is 30 mana. When it comes to a wizard's current mana levels, the same restrictions as the heal method apply.

The fire storm methods works similar to the fire bolt method. It is a form of attack, but instead of attacking a single target, it takes a list of creatures as an argument and attacks them all at once. A fire storm costs 50 mana to execute and causes one of two levels of damage to each creature. The lower level of damage is determined by a random number between 1 and 20 plus the wizard's speed value. The higher level of damage is determined by a random number between 5 and 20 plus the wizard's current arcana value. If the lower level damage value is greater than or equal to the wizards current arcana value, the lower level damage is applied, decreasing the health points of a creature by that value. Otherwise, the higher level damage is applied. The fire storm level and damage value is determined for each creature independently, meaning some creatures could be hit with a full fire storm and others with a mini. If a wizard has insufficient mana to execute a fire storm, a message is printed detailing how much mana they are short.

```python
# Exercise 3

# Task 1

class Wizard(Creature):

    def __init__(self, name, hp = 20, attack = 3, defence = 5, speed =5):
        self.mana = 100
        self.arcana = 10
        Creature.__init__(self, name, hp, attack, defence, speed, maxhp = 20)
        
    def __str__(self):
        return Creature.__str__(self) + ', Mana: {}, Arcana: {}'.format(self.maxhp, self.mana, self.arcana)
        
    def attack(self, target):
        Creature.attack(self, target)
        if self.mana + 20 <= 100:
            self.mana += 20
            print('Mana increased by 20.')
        elif self.mana == 100:
            print('No increase in mana levels. They are full!')
        else:
            self.mana = 100
            print('Mana returned to maximum (100).')
    
    def recharge(self):
        if self.mana + 30 <=100:
            self.mana += 30
            print('{} recharges with 30 mana points'.format(self.name))
        elif self.mana == 100:
            print('No need to recharge, Mana is full!')
        else:
            self.mana = 100
            print('Mana returned to maximum (100).')
        
    def fire_bolt(self, target):
        attackRoll = random.randint(1,20)
        targetDefence = int(target.abilities['Defence']) + int(target.abilities['Speed'])
        print('{} shoots a fire bolt at {}.'.format(self.name, target.name))
        if attackRoll > targetDefence:
            attackRoll = int(self.abilities['Attack']) + random.randint(1,self.arcana) + int(self.arcana/2)
            target.hp -= attackRoll
            print('Attack hits for {} damage!'.format(attackRoll))
            if self.mana + 10 <=100:
                self.mana += 10
                print('{} gains 10 mana points.'.format(self.name))
            elif self.mana == 100:
                print('No need to recharge, Mana is full!')
            else:
                self.mana = 100
                print('Mana returned to maximum (100).')
        else:
            print("Attack not successful.")
    
    def heal(self, target):
        if self.mana >= 20:
            self.mana -= 20
            healAmount = random.randint(0,8) + int(self.arcana/2)
            if target.hp + healAmount <= target.maxhp:
                target.hp += healAmount
                print('{} heals {} with {} health points. {}\'s health points increased to {}!'.format(self.name, target.name, healAmount, target.name, target.hp))
            else:
                target.hp = target.maxhp
                print('{} heals {} with {} health points. {}\'s health points are at maximum ({})!'.format(self.name, target.name, healAmount, target.name, target.hp))
        else:
            print('Insufficient Mana to Heal. You need {} more Mana points.'.format(20 - self.mana))
    
    def mass_heal(self, allies):
        if self.mana >= 30:
            self.mana -= 30
            healAmount = random.randint(0,10) + self.arcana
            if self.hp + healAmount <= self.maxhp:
                self.hp += healAmount
                print('{} heals themselves with {} health points. {}\'s health points increased to {}!'.format(self.name,healAmount, self.name,self.hp))
            else:
                self.hp = self.maxhp
                print('{} heals themselves with {} health points. {}\'s health points are at maximum ({})!'.format(self.name,healAmount, self.name,self.hp))
            for ally in allies:
                if ally.hp + healAmount <= ally.maxhp:
                    ally.hp += healAmount
                    print('{} heals {} with {} health points. {}\'s health points increased to {}! '.format(self.name, ally.name, healAmount, ally.name,ally.hp))    
                else:
                    ally.hp = ally.maxhp
                    print('{} heals {} with {} health points. {}\'s health points are at maximum ({})!'.format(self.name, ally.name, healAmount, ally.name, ally.hp))
        else:
            print('Insufficient Mana to Mass Heal. You need {} more Mana points.'.format(30 - self.mana))
                      
    def fire_storm(self, enemies):
        if self.mana >= 50:
            self.mana -= 50
            for enemy in enemies:
                smallDamageAmount = random.randint(1,20) + int(self.abilities['Speed'])
                fullDamageAmount = random.randint(5,20) + self.arcana
                if smallDamageAmount >= self.arcana:
                    enemy.hp -= smallDamageAmount
                    print('{}\'s mini fire storm hits {} with {} damage!'.format(self.name, enemy.name, smallDamageAmount))
                else:
                    enemy.hp -= fullDamageAmount
                    print('{}\'s full fire storm hits {} with {} damage!'.format(self.name, enemy.name, fullDamageAmount))
        else:
            print('Insufficient Mana to execute Fire Storm. You need {} more Mana points.'.format(50 - self.mana))
```

Here I create the battle class which simulates a battle between the allies and the enemies. By initialising the battle class we create a team of enemies and a team of allies. The enemies are represented in the enemies list and include a Goblin King, an Orc General, two Goblins, and two Orcs. The battle class has four methods, auto_select, select_target, player_turn, and start.

The auto select method is used by creatures during battle to randomly select a target from the opposing team. The method takes a list of creatures as it's argument. It checks the life of each creature in the list, adding creatures that are alive to the healthyTargets list. The final target object is then selected from the healthyTargets list. This means that only an alive creature can be selected as the target. The method returns the target object, which is then used as the argument in a creature's attack method. If no creatures in the list are alive, the method returns None.

The select target method is used by the wizard (player) to select a creature to attack or heal during battle. It takes a list of creatures as it's argument. It prints the list of creatures, detailing the index, name, health points, and max health points of each. Using a while loop, an index number is requested from the player. If the input is outside the index range, the program prints an error message and asks the player for input again. This will continue until the player provides an input within the index range. Once the player provides an accepted input, the target name is printed and the method returns the target object. The target object is then used as the argument in a wizard action method (attack, fire bolt, heal).

Every four rounds the player gets to take a turn. The player turn method determines how that turn functions. When the method is called it prints the player name, health points, max health points, and mana. It then prints the name, health points, and max health points of each ally, and a list of possible actions. Using a while loop, input is requested from the player. If the input doesn't match any of the possible actions, an error is printed. This is repeated until an acceptable input is given. Once an acceptable input is given, the selected action is printed, if neccessary, a target is selected using the select target method, and the required action method is called. The loop is then broken with a break statement. If the player decides to quit, the method returns 1. This value is used to trigger an exit mode in the start method.

The start method is used to begin the battle. It begins by determining the order in which creatures will take their turn. Fastest creatures go first. To set the order I created a dictionary called order and filled it with all of the creatures in the enemies and allies lists. The keys were creature objects. The values were the creature's speed value. I then used a lambda function to order the list by the values. Reverse = True returns a list in ascending order. To set no limit on the number of rounds, I set round = 1, and run the code while round > 0. The round starts by printing the round number, and then, using a for loop, it iterate through each creature in the dictionary. It first check that they are alive. If they are, it checks if they are an ally or an enemy. Based on this, it uses the auto select method to select a target. If the auto select method returns None, this means that an entire team has been killed and the opposing team has won. When this happens, a message indicating the winners is printed. A game over condition is also triggered, printing a message and a list of creatures that were killed. If there are alive creatures to choose from, the creature's turn method is called with the round number and target object. After the turn is complete, the target's life is checked. If it's 0, it has been killed and a message is printed to that effect. The names of killed creatures are appended to a list called 'dead'. This list of names is displayed at the end of each round. This process repeats for each creature in the dictionary. Once every alive creature has taken a turn, the round is complete. At the end of a round, if a team has not won, a message is printed indicating the end of the round, along with a list of creatures killed thus far. The round variable increments by 1 and a new round begins. Every 4 rounds, the player turn method is called allowing the player to interact in the battle.

```python
# Exercise 4 - Battle for Middle Earth

class Battle:

    def __init__ (self):
        self.goblinKing = GoblinKing('Goblin King')
        self.orcGen = OrcGeneral('Orc General')
        self.goblin1 = Goblin('Goblin One')
        self.goblin2 = Goblin('Goblin Two')
        self.orc1 = Orc('Orc One')
        self.orc2 = Orc('Orc Two') 
        self.enemies = [self.goblinKing, self.orcGen, self.goblin1, self.goblin2, self.orc1, self.orc2]
        self.fighter1 = Fighter('Fighter One')
        self.archer1 = Archer('Archer One')
        self.fighter2 = Fighter('Fighter Two')
        self.archer2 = Archer('Archer Two')
        self.allies = [self.fighter1, self.fighter2, self.archer1, self.archer2]
        self.player = Wizard('Wizard')
    
    def auto_select(self, target_list):
        healthyTargets = []
        for target in target_list:
            targetLife = target.check_life()
            if targetLife > 0:
                healthyTargets.append(target)
        if len(healthyTargets) == 0:
            return None
        else:
            target = random.choice(healthyTargets)
            return target

    def select_target(self, target_list):
        print()
        print('****** Target List ******')
        for target in target_list:
            print('{} Name: {}, HP: {}/{}'.format(target_list.index(target), target.name, target.hp, target.maxhp))
        print()

        while True:
            playerTarget = int(input('Which creature would you like to target? Enter it\'s index: '))
            if playerTarget > len(target_list)-1 or playerTarget < 0:
                print('Error. I don\'t recognise that index. Please try again...')
            else:
                print('You have chosen:', target_list[playerTarget].name)
                return target_list[playerTarget]
    
    def player_turn(self):
        wizard = self.player
        print('=========================')
        print('Player {} HP: {}/{} Mana: {}'.format(wizard.name, wizard.hp, wizard.maxhp, wizard.mana))
        print('Allies:')
        for ally in self.allies:
            print('{} HP: {}/{}'.format(ally.name, ally.hp, ally.maxhp))
        print('=========================')
        print('Actions: F: Attack R: Recharge Mana')
        print('Spells: 1. Heal 2. Firebolt 3. Mass Heal 4. Fire Storm\n To quit game type: Q')
        print('=========================')
        actionList = ['1','2','3','4','F','R','f','r','Q','q']
        while True:
            action = input('Enter input: ')
            if action not in actionList:
                print('Error. Please enter a valid option (1,2,3,4,R,F,Q)')
            if action == 'F' or action == 'f':
                print('You selected: Attack')
                target = self.select_target(self.enemies)
                wizard.attack(target)
                break
            if action == 'F' or action == 'r':
                print('You selected: Recharge Mana')
                wizard.recharge()
                break
            if action == '1':
                print('You selected: Heal')
                target = self.select_target(self.allies)
                wizard.heal(target)
                break
            if action == '2':
                print('You selected: Firebolt')
                target = self.select_target(self.enemies)
                wizard.fire_bolt(target)
                break
            if action == '3':
                print('You selected: Mass Heal')
                wizard.mass_heal(self.allies)
                break
            if action == '4':
                print('You selected: Fire Storm')
                wizard.fire_storm(self.enemies)
                break
            if action == 'Q' or action == 'q':
                print('You selected: Quit')
                return 1

    def start(self):
        order = {}
        for enemy in self.enemies:
            order[enemy] = enemy.abilities['Speed']
        for ally in self.allies:
            order[ally] = ally.abilities['Speed']
        order = sorted(order.items(), key=lambda x:x[1], reverse=True)
        dead = []

        round = 1
        while round > 0:
            print()
            print('****** Round {} ******'.format(round))
            for creature in range(len(order)):
                creature = order[creature][0]
                creatureLife = creature.check_life()
                if creatureLife > 0:
                    if creature in self.allies:
                        target = self.auto_select(self.enemies)
                        if target == None:
                            print()
                            print('**** The Allies have won! ****')
                            break
                    else:
                        target = self.auto_select(self.allies)
                        if target == None:
                            print()
                            print('**** The Enemies have won! ****')
                            break
                    print()
                    creature.turn(round, target)
                    targetLife = target.check_life()
                    if targetLife == 0:
                        print('{} has been killed!'.format(target.name))
                        dead.append(target.name)
            print()
            if target == None:
                print('*** GAME OVER **** \n\nThank you for playing.\n')
                print('The Fallen:', ', '.join(dead))
                break
            else:
                print('****** Round Complete ******')
                print('The Fallen:', ', '.join(dead))
                round +=1
                if round % 4 == 0:
                    target = self.player_turn()
                    if target == 1:
                        print('\n*** GAME OVER **** \n\nThank you for playing.\n')
                        print('The Fallen:', ', '.join(dead))
                        break
```
